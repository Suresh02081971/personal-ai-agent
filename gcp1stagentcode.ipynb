```python
import os
from langchain_openai import ChatOpenAI
from langchain_community.agent_toolkits.load_tools import load_tools
from langchain_classic import hub
#from langchain.agents import AgentExecutor, create_react_agent
from langchain_core.agents import AgentExecutor, create_react_agent



# 2. Setup LLM and Tools
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
tools = load_tools(["serpapi", "llm-math"], llm=llm)

# 3. Get the prompt template from LangChain Hub
# This is a standard prompt for ReAct agents
prompt = hub.pull("hwchase17/react")

# 4. Construct the Agent
agent = create_react_agent(llm, tools, prompt)

# 5. Create the Executor (the runtime)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True, handle_parsing_errors=True)

# 6. Run
agent_executor.invoke({"input": "Who is the current CEO of Tesla and what is their age raised to the power of 0.5?"})
```
